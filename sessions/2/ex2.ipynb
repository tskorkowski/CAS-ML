```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-100,100,1000)
y = np.linspace(-100,100,1000)

X,Y = np.meshgrid(x,y)
```

```python
def cont_plot(X,Y,*args):
    for arg in args:
        plt.contour(X,Y,arg)
    
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.show
```

```python
F = X**2 + Y**2 - 1
G = X*Y - 1
H = X**2 - Y**2

```

```python
def contour_grad(X,Y, func, grad_func, poi, axis):
    Z = func(X,Y)
    levels = 55
    scale = 1

    axis.contour(X,Y,Z, levels=levels)
    
    grad_start = poi
    grad_vect = grad_func(*poi)
    axis.arrow(x=grad_start[0], y=grad_start[1],
              dx=(grad_vect[0]-grad_start[0])/scale,
              dy=(grad_vect[1]/1-grad_start[1])/scale,
              head_width=2, head_length=1, fc='r', ec='r')
    return axis
```

```python
def f_f(x,y):
    return x**2 + y**2 - 1
def grad_f_f(x,y):
    return (2*x,2*y)

def f_g(x,y):
    return x*y - 1
def grad_f_g(x,y):
    return (y,x)

def f_h(x,y):
    return x**2 - y**2
def grad_f_h(x,y):
    return (2*x, 2*y)
```

```python
functions = [f_f, f_g, f_h]
gradients = [grad_f_f, grad_f_g, grad_f_h]
fig, ax = plt.subplots(1,3, figsize=(15, 4))
for i,axis in enumerate(ax):
    axis.set_xlabel('x')
    axis.set_ylabel('y')
    axis = contour_grad(X,Y, functions[i], gradients[i], (-25,15), axis)

plt.show()
```
